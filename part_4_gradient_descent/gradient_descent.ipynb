```python
X = np.array(((1, 3), (4,10)))
Y = np.array((5, 6))
X
```

```python
M = np.array([-1, 2])
B = np.array([1,1])
```

```python
def y_pred(X, M, B):
    """"This function finds y pred"""
    y_pred = np.matmul(X, M) + B
    return y_pred
```

```python
y_predicted = y_pred(X=X, M=M, B=B)
y_predicted.reshape(-1, 1)
```

```python
def errors(Y, Y_pred):
    """This function calculate the error"""
    sum_values = 0
    for i, j in zip(Y_pred, Y):
        sum_values += (i -j)**2
    mse = sum_values/(Y_pred.shape[0])
    return mse
```

```python
mse = errors(Y=Y, Y_pred=y_predicted)
print(mse)
```
